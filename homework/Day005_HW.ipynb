```python
x=json.loads(r.text)
print(len(x))
##30筆資料
```

```python
print (x[0].keys())
##包含哪些欄位
```

```python
for t in x:
    print(t['title'])
##標題
```

```python
for c in x:
    print(c['createdAt'])
##貼文時間
```

```python
for comment in x:
    print (comment['commentCount'])
##留言個數
```

```python
##熱門文章
for hot in x:
    sum+=hot['commentCount']
print ("平均留言人數：",'',sum/len(x))
for like in x:
    sum+=like['likeCount']
print ("平均按讚人數",'', sum/len(x))
```
